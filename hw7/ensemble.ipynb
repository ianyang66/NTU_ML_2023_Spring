```python
# Reference:　https://docs.python.org/zh-tw/3/library/collections.html
import pandas as pd
from collections import Counter

# path_6 = ''
path_1 = './result-tok-v6.csv'
path_2 = './result-tok-v7.csv'
path_3 = './result-tok-v4.csv'
path_4 = './result-tok-v3.csv'
path_5 = './result-tok-v2.csv'
path_6 = './result1-mlhw7-v2.csv'
path_7 = './result1-mlhw7-v3.csv'
path_8 = './result1-mlhw7-v4.csv'
path_9 = './result1-mlhw7-v6.csv'
path_10 = './result1-mlhw7-v7.csv'
path_11 = './result-v19-ian.csv'
path_12 = './result-v11-ian.csv'
path_13 = './result-v18-ian.csv'
path_14 = './result-t-v4.csv'
path_15 = './result-t-v5.csv'
path_16 = './result-epo2+1-v6.csv'
path_17 = './result-pertmrc-epo3.csv'

def post_process(answer):
    if '《' in answer and '》' not in answer:
            answer = answer + '》'
    if '《' not in answer and '》' in answer:
        answer = '《' + answer
    if '「' in answer and '」' not in answer:
        answer = answer + '」'
    if '」' in answer and '「' not in answer:
        answer = answer + '」'
    answer = answer.replace('[UNK]','')
    answer = answer.replace('[PAD]','')
    answer = answer.replace('[SEP]','')
    return answer

def append_in(path):
    arr = []
    arr0 = []
    with open(path) as f1:
        for lines in f1:
            if (lines != 'ID,Answer\n'):
                ans = post_process(lines.split(',')[-1].split('\n')[0])
                print(ans)
                arr.append(ans)
                arr0.append(lines.split(',')[0].split('\n')[0])
    return arr, arr0

arr_1, arr0_1 = append_in(path_1)

arr_2, arr0_2 = append_in(path_2)
arr_3, arr0_3 = append_in(path_3)
arr_4, arr0_4 = append_in(path_4)
arr_5, arr0_5 = append_in(path_5)
arr_6, arr0_6 = append_in(path_6)
arr_7, arr0_7 = append_in(path_7)
arr_8, arr0_8 = append_in(path_8)
arr_9, arr0_9 = append_in(path_9)
arr_10, arr0_10 = append_in(path_10)
arr_11, arr0_11 = append_in(path_11)
arr_12, arr0_12 = append_in(path_12)
arr_13, arr0_13 = append_in(path_13)
arr_14, arr0_14 = append_in(path_14)
arr_15, arr0_15 = append_in(path_15)
arr_16, arr0_16 = append_in(path_16)
arr_17, arr0_17 = append_in(path_17)

pred = []
for i in range(len(arr_1)):
    temp = []
    temp.append(arr_1[i])
    temp.append(arr_2[i])
    temp.append(arr_3[i])
    temp.append(arr_4[i])
    temp.append(arr_5[i])
    temp.append(arr_6[i])
    temp.append(arr_7[i])
    temp.append(arr_8[i])
    temp.append(arr_9[i])
    temp.append(arr_10[i])
    temp.append(arr_11[i])
    temp.append(arr_12[i])
    temp.append(arr_13[i])
    temp.append(arr_14[i])
    temp.append(arr_15[i])
    temp.append(arr_16[i])
    temp.append(arr_17[i])
    classification = Counter(temp)
    pred.append(classification.most_common(1)[0][0])

# for i in range(len(arr0_1)):
#     temp2 = []
#     temp2.append(arr0_1[i])
# print(temp2)
# print(arr0_1)
df = pd.DataFrame()
df["ID"] = arr0_1
print(df)
df["Answer"] = pred
df.to_csv("result-ensemble-2-17-post.csv", index = False)
```
